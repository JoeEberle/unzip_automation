```python
def launch_application(app_to_launch_command):
    ''' Launch any application or local executable ''' 
    # Run the command using subprocess.call
    return_code = subprocess.call(app_to_launch_command, shell=True)
    
    if return_code == 0:
        return f"Command '{app_to_launch_command}' executed successfully." 
    else:
        return f"Command '{app_to_launch_command}' failed with exit code {return_code}." 
```
